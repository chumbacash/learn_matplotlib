````python
#Matplotlib Pyplot

#Pyplot
'''
Most of the matplotlib utilities lies under pyplot submodule, and usually
imported under the plt alias:
'''
#Example: Draw a line in diagram from position (0,0) to position (6,250):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
````

````python
#Matplotlib plotting

#Plotting x and y points
'''
The plot() function is used to draw points (markers) in a diagram.
by default, the plot() function draws a line from point to point.

takes parameters:
- array containing points on the x-axis
- array containing points on the y-axis

example if we need to plot (1, 3) and (8, 10), we have to pass two arrays
[1, 8] and [3, 10] to the plot function.
'''

#Example: Draw a line in a diagram from position (1, 3) to position (8, 10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()

#Plotting without a line which means it uses 'rings' parameter is 'o'

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
````

````python
#Plotting multiple points
'''
You can plot as many points as you like, just make sure you have the same
number of points in both axis.
'''

#Example:
'''Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1)
and finally to position (8, 10):'''

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()

#Default X-points
#Example: Plotting without x-points:

xpoints = np.array([3, 8, 1, 10, 5,7])

plt.plot(ypoints)
plt.show()
````

````python
#Markers:
'''
You can use keyword marker to emphasize each point with a specified
marker.
'''
#Example: Mark each point with a circle:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')

plt.show()

plt.plot(ypoints, marker = '*')
````

````python
#Format Strings
'''
Use the shortcut string notation parameter to specify the marker.
This parameter is also called fmt written as marker|line|color
'''
#Example: Mark each point with circle:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:b') #b represent the color
plt.show()

# 'r'	Red
# 'g'	Green
# 'b'	Blue
# 'c'	Cyan
# 'm'	Magenta
# 'y'	Yellow
# 'k'	Black
# 'w'	White
````

````python
#Marker Size - We can use ms, to set size of markers.

#Example: Set size of markers to 20:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*', ms = 20)
plt.show()
````

````python
#Marker Color - mec to set the color edge of the markers:

#Exampe: Set the EDGE color to red:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*', ms = 20, mec = 'g')
plt.show()
````

````python
#Markerface Color - mfc to set the color indide edge markers

#Example: Set the FACE color to red:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*', ms = 20, mfc = 'w') #Set to white
plt.show()
````

````python
#Matplotlib Line:

#Linestyle - ls is used to change the style of the plotted line

#Example: Use a dotted line:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = 'dotted')
plt.show()

plt.plot(ypoints, ls = 'dashed') #Dashed

#Also there are other linstyles to be applied. 'solid', 'dashdot', 'none'
````

````python
#Line color - Using argument color or c to set line color

#Example: Set the color to red:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r', ls = 'dashed')
plt.show()

#Line Width
'''
You can use linewith or lw to change the width of the line.
The value is a floating number, in points:
'''

#Example: Plot with 20.5pt wide line:

#Using the above points
plt.plot(ypoints, lw = '20.5')
plt.show()
````

````python
#Multiple Lines: We can do that simply by adding more plt.plot() function:

#Example: Draw two line specifying plt.plot() function for each line.

import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.show()

#Example2: Draw two lines by specifying the x- and y-point values for both line

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
````

````python
#Matplotlib Labels and Title

#Create Labels for a Plot
'''
With pyplot, you can use the xlabel() and ylabel() functions to set
a label for the x- and y-axis
'''

#Example: Add labels to the x- and y -axis:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()

#Create a title for a Plot - Using title() function to set a title for the plot

#Example: Add a plot title and labels for the x- and y -axis


import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data", color = 'g')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
````

````python
#Set Font properties for Title and Labels
'''
You can use the fontdict parameter in xlabel(), ylabels(), and title() to
set font properties for the tile and labels.
'''

#Example: Set font properties for the title and labels:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family': 'serif', 'color':'darkred', 'size':10}
plt.plot(x, y)

plt.title("Sports Watch Data", fontdict= font1)
plt.xlabel("Average Pulse", fontdict= font2)
plt.ylabel("Calorie Burnage", fontdict= font2)

plt.show()
````

````python
# Position The Title:
'''
You can use the loc parameter in title() to position.
Legal values are: 'left', 'right', and 'center'. Default user is 'center'
'''

#Example: Position the title to the left:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
````

````python
#Adding Grid Lines
'''
With Pyplot, you can use the grid() function to add grid lines to the plot
'''

#Example: Add grid lines to the plot:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.grid()
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()

#Specify which grid lines to specify
#Use axis parameter in the grid() function also legal values 'x','y' and 'both'

#Example: Display only grid lines for the x-axis

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.grid(axis = "x")
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()

#Example: Display only grid lines for the y-axis:

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.grid( axis = 'y')
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()

'''
You can also set the line properties of the grid,
like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).
'''
````

````python
#Matplotlib Subplot

#Display Multiple plots: with subplot() func you can draw multiple plots

#Example Draw 2 plots:

import matplotlib.pyplot as plt
import numpy as np

#plot1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)

#plot2

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
````

###Subplot Function
You can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```
plt.subplot(1, 2, 1)
```

the figure has 1 row, 2 columns, and this plot is the first plot.



```
plt.subplot(1, 2, 2)
```
the figure has 1 row, 2 columns, and this plot is the second plot.

So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:

````python
#Example: Draw 2 plots on top of each other

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
````

````python
#Matplotlib Scatter

#Creating Scatter Plots
'''
With Pyplot, you can use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation.It needs  two arrays
of the same length, one for the values of the axis, and one for values on the
y-axis:
'''

#Example: A simple scatter plot:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
````

````python
#Color each dot - set a specific color for each dot using c argument

#Example: Set your own color of the markers

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange",
                   "purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c = colors)

plt.show()
````

````python
#ColorMap is called 'viridis'
'''
How to Use the ColorMap
You can specify the colormap with the keyword argument cmap with the value
the colormap, this case 'viridis' which is one of the built-in colormaps
available in Matplotlib.

In addition you have to create an array with values (from 0 to 100),
one value for each point in the scatter plot:
'''

#Example: Create a color array, and specify a colormap in the scatter plot:

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y , c = colors, cmap = 'viridis')

plt.colorbar() #Include the actual colormap:
plt.show()
````

###size
You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

````python
#Example:Set your own size for the markers:

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
````

##Alpha

You can adjust the transparency of the dots with the alpha argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

````python
#Example: Set your own size for the markers

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
````

##Combine Color Size and Alpha

You can combine a colormap with different sizes of the dots. This is best visualized if the dots are transparent:

````python
#Example: Create a random arrays with 100 values for x and y points, color and sizes

x = np.random.randint(100, size=100)
y = np.random.randint(100, size=100)
colors = np.random.randint(100, size=100)
sizes = 10* np.random.randint(100, size=100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
````

##Matplotlib Bars
###Creating Bars
With Pyplot, you can use the bar() function to draw bar graphs:


````python
#Example: Draw 4 bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()

#Bar2

x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
````

````python
#Horizontal Bars use the barh() functionto display it horizontally

#Example: Draw 4 horizontal bars:

x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])

plt.barh(x,y)
plt.show()
````

###Bar Color
The bar() and barh() take the keyword argument color to set the color of the bars:

````python
#Draw 4 red bars

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "pink") #setting color you can use hexadecimal colors
plt.show()
````

####Bar Width
The bar() takes the keyword argument width to set the width of the bars:

The default width value is **0.8**

Note: For horizontal bars, use height instead of width.

````python
#Example: Draw 4 very thin bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1) #For barh( height = 0.1)

plt.show()
````

#Matplotlib Pie Charts

##Creating pie charts
With Pyplot, you can use the pie() function to draw pie charts:

````python
#Example: A simple pie chart:

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()

#Plotting of the first wedge starts from the x-axis and moves counterclockwise:
````

##Labels

Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge:

````python
#Example: A simple pie chart:

y = np.array([35, 25, 25, 15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Dates']

plt.pie(y, labels = mylabels)
plt.show()
````

###Start Angle
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a `startangle` parameter.

The `startangle` parameter is defined with an angle in degrees, default angle is 0:

````python
#Example - Start the first wedge at 90 degrees:

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
````

###Explode
Maybe you want one of the wedges to stand out? The `explode` parameter allows you to do that.

The `explode` parameter, if specified, and not `None`, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

````python
#Example: Pull the "Apples"wedge 0.2 from the center of the pie

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
````

####Shadow
Add a shadow to the pie chart by setting the `shadows` parameter to `True`:

````python
#Example:Add a shadow:

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
````

####Colors
You can set the color of each wedge with the `colors` parameter.

The **colors** parameter, if specified, must be an array with one value for each wedge:

````python
#Example: Specify a new color for each wedge:

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
````

####Legend
To add a list of explanation for each wedge, use the `legend()` function:

````python
#Example: Add Legend

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
````

````python
#Example: Add a legend with header:

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "My Fruits:")
plt.show()
````
